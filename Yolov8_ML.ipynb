```python
import csv
tsv_file = "/content/gdrive/MyDrive/Yolo_test/output/vaild.txt"
```

```python
with open(tsv_file,'r') as f:
  reader = csv.DictReader(f,delimiter = ' ')
  tsv_data = list(reader)
print(reader.fieldnames)
```

```python
import pandas as pd
df = pd.DataFrame(tsv_data)
```

```python
import shutil
dest_1 = '/content/gdrive/MyDrive/Yolo_test/out_dangerous'
dest_2 = '/content/gdrive/MyDrive/Yolo_test/out_non_dangerous'
for r, row in df.iterrows():
  img = row['image_path'].strip(":")
  if row['cls1'] == 'no_detections':
    shutil.copy(img, dest_2)
  else:
    shutil.copy(img, dest_1)
```
